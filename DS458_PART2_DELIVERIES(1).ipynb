The process was carried out in several steps:

**1. Missing Values:**
"TotalCharges" was the only column with missing values. Which got filled with the mean of that column.

**2. Outliers:**
I tried 2 techniques for outliers; Z-Score and Quantile. Both showed no outliers in the data.

**3. Categorical Data Encoding:**
I used 2 techniques; LabelEncoder and Dummy Encoding. LabelEncoder data is what I ended up working with since it gave me better results. And it was also used to get the correlation matrix between the entire features.

**4. Feature Engineering:**
I engineered 7 features total: Tenure Group, MonthlyChargesGroup, TotalChargesGroup, Has Phone And Internet, Service Reliability, Number of Services, Changing Values of Contract to Month.
***Important takeaways:***
- The first 3 features are Numerical. The engineering here was done for the encoding part. I ended up keeping the charges as numerical (it gave better results), and the 'tenure' feature got divided into 5 groups: ['0-1 year', '1-2 years', '2-4 years', '4-5 years', '5+ years']
- Changing Values of Contract to Month: Changed from categorical to numerical which gave better results.
(Not a surprize since ML models prefer numerical sometimes)
- Service Reliability (mix of OnlineSecurity, TechSupport, OnlineBackup, and DeviceProtection) was a great addition since it encapsulates the overall level of service a customer is receiving. It helps identify customers who may feel underserved, allowing for targeted retention efforts.

**5. Feature Scaling:**
The numerical features (MonthlyCharges, TotalCharges) got normalized for better Machine Learning training using Standardization.
